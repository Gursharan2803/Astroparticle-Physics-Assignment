```python
import numpy as np
from corsikaio import CorsikaParticleFile, as_dict
import matplotlib.pyplot as plt
```

```python
with CorsikaParticleFile('DAT000002',thinning=True) as f:
    print("Run Number",f.run_header['run_number'])
    print("CORSIKA Version",f.version)

    for e in f:
        print("Energy =",e.header['total_energy'],"GeV")
        print("Minimum Theta =",e.header['theta_min']) 
        print("Zenith angle =",e.header['zenith']*360/(2*np.pi),"degrees")
        print("Particle ID:",e.particles['particle_description'])
        
        print("Thinning Levels:",e.particles['thinning_level'])
        plt.scatter(e.particles['x'], e.particles['y'])
        plt.show()
```
